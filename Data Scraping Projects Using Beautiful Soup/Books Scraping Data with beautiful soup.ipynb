### Scrape 51 pages of the book website 

```python
books=[]
for i in range(1,51) :
    url=f'https://books.toscrape.com/catalogue/page-{i}.html'
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html.parser')
    lists=soup.find_all('ol',class_='row')
    articles=lists[0].find_all('article',class_="product_pod")
    for i in articles :
        image=i.find('img')
        title=image.attrs['alt']
        src= "https://books.toscrape.com/"+image.attrs['src']
        star=i.find('p')
        star=star['class']
        price=i.find('p',class_='price_color').text
        availability=i.find('p',class_='instock availability').text.replace('\n','').strip()
        books.append([title,src,star[1],price,availability])
```

### Save data to dataframe 

```python
df=pd.DataFrame(books,columns=['Title','Source','Stars','Price','Availability'])
```

### Save data to csv

```python
df.to_csv('booksscraping.csv',index=False,header=True)
```
